```python
import requests
import re
import uniprot
import pp


# Retrieve FASTA file from the website
url = 'https://rest.uniprot.org/uniprotkb/stream?format=fasta&query=%28pyrococcus%20AND%20%28taxonomy_id%3A2261%29%29'
all_fastas = requests.get(url).text


# Compile the regular expression
pattern = re.compile(r".*OX=186497.*")

# Initialize an empty list to store the lines
lines = []

# Iterate over the lines of the FASTA file
for line in all_fastas.splitlines():
    if re.search(pattern, line):
        lines.append(line)
# print(lines)
pp(lines)
```
