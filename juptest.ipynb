```python
import pandas as pd
import seaborn as sns

# Read in data
df = pd.read_excel("D:/1_DATA/Rcoding/Heatmap_try/SEC_ribosomal_prot.xlsx")

```

```python
print(df)
```

```python
# Extract short form of Fasta headers
df['fastaShort'] = df['Fasta_headers'].str.replace(r' OS=.*', '')

# Normalize data on columns
df_normalized = df[df.columns[6:]]
df_normalized = (df_normalized - df_normalized.min()) / (df_normalized.max() - df_normalized.min())

# Merge normalized data back into original dataframe
df_normalized = pd.concat([df[df.columns[:6]], df_normalized], axis=1)

# Drop rows with missing values
df_normalized.dropna(axis=0, how='any', inplace=True)

# Reshape data for heatmap
df_long = pd.melt(df_normalized, id_vars=['Fasta_headers', 'fastaShort', 'Mol_weight', 'Sequence_coverage', 'Intensity', 'iBAQ', 'iBAQ_peptides'], var_name='Fractions', value_name='iBAQ_values')

# Create heatmap
sns.heatmap(df_long.pivot_table(index='Fractions', columns='fastaShort', values='iBAQ_values'))
```
